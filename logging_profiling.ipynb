```python
import cProfile, pstats, pandas as pd


def do_nothing():
    print('I am as lazy as you. I do nothing.')


def profile_function(function , sort_by = 'tottime',  return_stats = False):
    # sort by ['Ncalls', 'Ccalls', 'ToTtime', 'CumTime', 'Callers']
    
    def wrapped_function(*args, **kwargs):
        with cProfile.Profile() as profiler:
            output =  function(*args, **kwargs)
            
        print(pstats.Stats(profiler).sort_stats(sort_by).print_stats())

        return output, stats if return_stats else output
        
    return wrapped_function
    
f = profile_function(do_nothing, return_stats=True)
_, stats =f()
stats.to_csv('stats.csv')
```
