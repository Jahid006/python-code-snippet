```python
import shutil

f=open('data.txt','r')
f1=open('data1.txt','w')
 
# Syntax: shutil.copyfileobj(src,dst)
shutil.copyfileobj(f,f1)
 
f.close()
f1.close()
```

```python
#shutil.copy(src,dst)
shutil.copy('data.txt','data1.txt')
```

```python
#shutil.copy2(src,dst)
#copy(src,dst) and copy2(src,dst) functions are almost the same but copy2(src,dst) also copies the metadata of the source file
shutil.copy2('data.txt','data1.txt')
```

```python
shutil.move(src,dst)
shutil.copyfile(src,dst)
shutil.copytree(src,dst)
```

```python
#By using shutil.rmtree(), we can delete any folder,file or directory. Letâ€™s delete the folder Python.

shutil.rmtree('path')
```

```python
#function returns the disk usage statistics about the given path names as tuple with attributes total 
# which is total amount of memory, used which is used space and free which is free space in bytes
shutil.disk_usage('path') 
```

```python
#function returns the path to an executable application which would run if the given command cmd was called
shutil.which() 
```
